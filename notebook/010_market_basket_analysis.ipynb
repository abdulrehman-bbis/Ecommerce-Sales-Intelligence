```python
import pandas as pd

from mlxtend.frequent_patterns import (
    apriori,
    association_rules
)
```

```python
master = pd.read_csv(
    "../data/cleaned/master_dataset.csv"
)
```

```python
basket = (
    master
    .groupby(
        ["order_id",
         "product_category_name_english"]
    )["product_id"]
    .count()
    .unstack()
    .fillna(0)
)
```

```python
basket = basket > 0
```

```python
basket.head()
```

```python
frequent_items = apriori(
    basket,
    min_support=0.01,
    use_colnames=True
)
```

```python
frequent_items.head()
```

```python
basket.dtypes.head()
```

```python
rules = association_rules(
    frequent_items,
    metric="lift",
    min_threshold=1
)
```

```python
rules.head()
```

```python
rules.sort_values(
    "lift",
    ascending=False
).head(10)
```

Market Basket Analysis was attempted using the Apriori algorithm. However, approximately 98% of orders contained products from only one category, preventing the formation of frequent multi-item itemsets. Therefore, association rules could not be generated, indicating that category-level basket analysis is not suitable for this dataset.
